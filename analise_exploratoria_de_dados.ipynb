```python
# ============================================================================
# DESAFIO T√âCNICO ‚Äì AN√ÅLISE EXPLORAT√ìRIA E MODELAGEM COM BASE DE DADOS
# Dataset: dados.csv (sem informa√ß√µes pr√©vias)
# Objetivo: EDA + An√°lise Estat√≠stica + Machine Learning
# Autor: Ednilton Curt Rauh - edrauh@gmail.com
# Data: 22/06/2025
# ============================================================================

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier, RandomForestRegressor
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import silhouette_score, classification_report, r2_score
import warnings
warnings.filterwarnings('ignore')

# Configura√ß√£o visual
plt.style.use('seaborn-v0_8')
sns.set_palette("husl")
plt.rcParams['figure.figsize'] = (12, 8)

print(" üéØ DESAFIO T√âCNICO - AN√ÅLISE COMPLETA DE DADOS")
print("="*50)
print("üìã ENTREG√ÅVEIS:")
print("‚úÖ 1. An√°lise Explorat√≥ria de Dados (EDA)")
print("‚úÖ 2. An√°lise Estat√≠stica")
print("‚úÖ 3. Aplica√ß√£o de T√©cnicas de Machine Learning")
print("‚úÖ 4. Documenta√ß√£o e Clareza")
print("="*50)

```

**SE√á√ÉO 1:** AN√ÅLISE EXPLORAT√ìRIA DE DADOS (EDA)

```python
# 1.1 Carregamento e Primeira Inspe√ß√£o
print("\nüîπ 1.1 CARREGAMENTO E PRIMEIRA INSPE√á√ÉO")
dados_tabela = pd.read_csv("/content/drive/MyDrive/processo_seletivo_senai/dados.csv")

print(f"üìê Dimens√µes: {dados_tabela.shape[0]} linhas √ó {dados_tabela.shape[1]} colunas")
print(f"üìã Colunas: {list(dados_tabela.columns)}")

# Exibir amostra dos dados
print("\nüìã AMOSTRA DOS DADOS:")
display(dados_tabela.head())

# 1.2 An√°lise de Qualidade dos Dados
print("\nüîπ 1.2 QUALIDADE DOS DADOS")
print(f"üîç Valores ausentes: {dados_tabela.isnull().sum().sum()}")
print(f"üîç Duplicatas: {dados_tabela.duplicated().sum()}")

# Tipos de dados
print("\nüìä TIPOS DE DADOS:")
for col in dados_tabela.columns:
    print(f"   {col}: {dados_tabela[col].dtype} ({dados_tabela[col].nunique()} valores √∫nicos)")

# 1.3 Estat√≠sticas Descritivas
print("\nüîπ 1.3 ESTAT√çSTICAS DESCRITIVAS")
display(dados_tabela.describe())

# 1.4 Distribui√ß√µes das Vari√°veis
print("\nüîπ 1.4 DISTRIBUI√á√ïES DAS VARI√ÅVEIS")
# Histogramas
fig, axes = plt.subplots(4, 4, figsize=(20, 16))
axes = axes.ravel()

for i, col in enumerate(dados_tabela.columns):
    if i < len(axes):
        axes[i].hist(dados_tabela[col], bins=20, alpha=0.7, edgecolor='black')
        axes[i].set_title(f'Distribui√ß√£o - {col}')
        axes[i].set_xlabel(col)
        axes[i].set_ylabel('Frequ√™ncia')

# Remove subplot extra se houver
if len(dados_tabela.columns) < len(axes):
    axes[-1].remove()

plt.tight_layout()
plt.show()

# 1.5 Box plots para outliers
print("\nüîπ 1.5 DETEC√á√ÉO DE OUTLIERS")
fig, axes = plt.subplots(4, 4, figsize=(20, 12))
axes = axes.ravel()

for i, col in enumerate(dados_tabela.columns):
    if i < len(axes):
        axes[i].boxplot(dados_tabela[col])
        axes[i].set_title(f'Box Plot - {col}')
        axes[i].set_ylabel(col)

if len(dados_tabela.columns) < len(axes):
    axes[-1].remove()

plt.tight_layout()
plt.show()
```

üìà SE√á√ÉO 2: AN√ÅLISE ESTAT√çSTICA

```python
print("\n" + "="*50)
print("üìà SE√á√ÉO 2: AN√ÅLISE ESTAT√çSTICA")
print("="*50)

# 2.1 Matriz de Correla√ß√£o
print("\nüîπ 2.1 AN√ÅLISE DE CORRELA√á√ÉO")
correlation_matrix = dados_tabela.corr()

plt.figure(figsize=(12, 10))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', center=0,
            fmt='.2f', square=True, linewidths=0.5)
plt.title('Matriz de Correla√ß√£o entre Vari√°veis', fontsize=16, fontweight='bold')
plt.tight_layout()
plt.show()

# Identificar correla√ß√µes fortes
print("üîç CORRELA√á√ïES MAIS FORTES:")
correlacoes_fortes = []
for i in range(len(correlation_matrix.columns)):
    for j in range(i+1, len(correlation_matrix.columns)):
        corr_val = correlation_matrix.iloc[i,j]
        if abs(corr_val) > 0.5:  # Correla√ß√µes moderadas a fortes
            correlacoes_fortes.append({
                'Var1': correlation_matrix.columns[i],
                'Var2': correlation_matrix.columns[j],
                'Correlacao': corr_val
            })

for item in sorted(correlacoes_fortes, key=lambda x: abs(x['Correlacao']), reverse=True)[:10]:
    strength = "üî¥ FORTE" if abs(item['Correlacao']) > 0.7 else "üü° MODERADA"
    print(f"   {item['Var1']} ‚Üî {item['Var2']}: {item['Correlacao']:.3f} ({strength})")

# 2.2 An√°lise de Componentes Principais (PCA)
print("\nüîπ 2.2 AN√ÅLISE DE COMPONENTES PRINCIPAIS (PCA)")

# Padronizar dados para PCA
scaler = StandardScaler()
dados_padronizados = scaler.fit_transform(dados_tabela)

# Aplicar PCA
pca = PCA()
pca_result = pca.fit_transform(dados_padronizados)

# An√°lise da vari√¢ncia explicada
plt.figure(figsize=(15, 5))

plt.subplot(1, 3, 1)
plt.plot(range(1, len(pca.explained_variance_ratio_) + 1),
         pca.explained_variance_ratio_, 'bo-')
plt.title('Vari√¢ncia Explicada por Componente')
plt.xlabel('Componente Principal')
plt.ylabel('Vari√¢ncia Explicada')
plt.grid(True, alpha=0.3)

plt.subplot(1, 3, 2)
cumsum_var = np.cumsum(pca.explained_variance_ratio_)
plt.plot(range(1, len(cumsum_var) + 1), cumsum_var, 'ro-')
plt.title('Vari√¢ncia Explicada Acumulada')
plt.xlabel('Componente Principal')
plt.ylabel('Vari√¢ncia Acumulada')
plt.axhline(y=0.8, color='green', linestyle='--', alpha=0.7, label='80%')
plt.axhline(y=0.95, color='orange', linestyle='--', alpha=0.7, label='95%')
plt.legend()
plt.grid(True, alpha=0.3)

# N√∫mero de componentes necess√°rios
n_comp_80 = np.argmax(cumsum_var >= 0.8) + 1
n_comp_95 = np.argmax(cumsum_var >= 0.95) + 1

plt.subplot(1, 3, 3)
# Visualiza√ß√£o 2D dos primeiros componentes
pca_2d = PCA(n_components=2)
pca_2d_result = pca_2d.fit_transform(dados_padronizados)
plt.scatter(pca_2d_result[:, 0], pca_2d_result[:, 1], alpha=0.6)
plt.title('Dados no Espa√ßo dos 2 Primeiros PCs')
plt.xlabel(f'PC1 ({pca_2d.explained_variance_ratio_[0]:.1%} da vari√¢ncia)')
plt.ylabel(f'PC2 ({pca_2d.explained_variance_ratio_[1]:.1%} da vari√¢ncia)')
plt.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

print(f"üìä Para 80% da vari√¢ncia: {n_comp_80} componentes")
print(f"üìä Para 95% da vari√¢ncia: {n_comp_95} componentes")

# 2.3 An√°lise de Outliers Detalhada
print("\nüîπ 2.3 AN√ÅLISE DETALHADA DE OUTLIERS")

def detectar_outliers_iqr(data, coluna):
    Q1 = data[coluna].quantile(0.25)
    Q3 = data[coluna].quantile(0.75)
    IQR = Q3 - Q1
    limite_inferior = Q1 - 1.5 * IQR
    limite_superior = Q3 + 1.5 * IQR
    outliers = data[(data[coluna] < limite_inferior) | (data[coluna] > limite_superior)]
    return len(outliers), limite_inferior, limite_superior

outliers_summary = {}
total_outliers = 0

print("üìã RELAT√ìRIO DE OUTLIERS (M√©todo IQR):")
for col in dados_tabela.columns:
    n_outliers, lim_inf, lim_sup = detectar_outliers_iqr(dados_tabela, col)
    outliers_summary[col] = n_outliers
    total_outliers += n_outliers

    if n_outliers > 0:
        porcentagem = (n_outliers / len(dados_tabela)) * 100
        print(f"   {col}: {n_outliers} outliers ({porcentagem:.1f}%) - Limites: [{lim_inf:.2f}, {lim_sup:.2f}]")

print(f"\nüî¢ TOTAL: {total_outliers} outliers identificados")

# 2.4 An√°lise de Clustering Inicial
print("\nüîπ 2.4 AN√ÅLISE DE CLUSTERING INICIAL")

# M√©todo do cotovelo para determinar k √≥timo
inertias = []
silhouette_scores = []
K_range = range(2, 11)

for k in K_range:
    kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
    kmeans.fit(dados_padronizados)
    inertias.append(kmeans.inertia_)
    labels = kmeans.labels_
    silhouette_avg = silhouette_score(dados_padronizados, labels)
    silhouette_scores.append(silhouette_avg)

# Visualizar an√°lise de k
plt.figure(figsize=(15, 5))

plt.subplot(1, 3, 1)
plt.plot(K_range, inertias, 'bo-')
plt.title('M√©todo do Cotovelo')
plt.xlabel('N√∫mero de Clusters (k)')
plt.ylabel('In√©rcia')
plt.grid(True, alpha=0.3)

plt.subplot(1, 3, 2)
plt.plot(K_range, silhouette_scores, 'ro-')
plt.title('Silhouette Score por k')
plt.xlabel('N√∫mero de Clusters (k)')
plt.ylabel('Silhouette Score')
plt.grid(True, alpha=0.3)

# Encontrar melhor k
melhor_k = K_range[np.argmax(silhouette_scores)]
print(f"üéØ Melhor k (Silhouette Score): {melhor_k}")

# Aplicar clustering com melhor k
kmeans_final = KMeans(n_clusters=melhor_k, random_state=42, n_init=10)
cluster_labels = kmeans_final.fit_predict(dados_padronizados)

plt.subplot(1, 3, 3)
scatter = plt.scatter(pca_2d_result[:, 0], pca_2d_result[:, 1],
                     c=cluster_labels, cmap='viridis', alpha=0.7)
plt.colorbar(scatter)
plt.title(f'Clusters (k={melhor_k}) no Espa√ßo PCA')
plt.xlabel(f'PC1 ({pca_2d.explained_variance_ratio_[0]:.1%})')
plt.ylabel(f'PC2 ({pca_2d.explained_variance_ratio_[1]:.1%})')

plt.tight_layout()
plt.show()

# Distribui√ß√£o dos clusters
cluster_counts = pd.Series(cluster_labels).value_counts().sort_index()
print("\nüìä DISTRIBUI√á√ÉO DOS CLUSTERS:")
for cluster, count in cluster_counts.items():
    porcentagem = (count / len(cluster_labels)) * 100
    print(f"   Cluster {cluster}: {count} amostras ({porcentagem:.1f}%)")
```

ü§ñ SE√á√ÉO 3: APLICA√á√ÉO DE T√âCNICAS DE MACHINE LEARNING

```python
# 3.1 Prepara√ß√£o dos Dados
print("\nüîπ 3.1 PREPARA√á√ÉO DOS DADOS PARA ML")

# Adicionar labels de cluster ao dataset
dados_com_clusters = dados_tabela.copy()
dados_com_clusters['Cluster'] = cluster_labels

# Estrat√©gia 1: Classifica√ß√£o (Predizer clusters)
print("\nüéØ ESTRAT√âGIA 1: CLASSIFICA√á√ÉO DE CLUSTERS")
X_classificacao = dados_tabela  # Features originais
y_classificacao = cluster_labels  # Target: clusters

# Dividir dados
X_train_cls, X_test_cls, y_train_cls, y_test_cls = train_test_split(
    X_classificacao, y_classificacao, test_size=0.3, random_state=42,
    stratify=y_classificacao
)

# Padronizar features
scaler_cls = StandardScaler()
X_train_cls_scaled = scaler_cls.fit_transform(X_train_cls)
X_test_cls_scaled = scaler_cls.transform(X_test_cls)

# 3.2 Modelos de Classifica√ß√£o
print("\nüîπ 3.2 MODELOS DE CLASSIFICA√á√ÉO")

modelos_classificacao = {
    'Random Forest': RandomForestClassifier(n_estimators=100, random_state=42),
    'Logistic Regression': LogisticRegression(random_state=42, max_iter=1000)
}

resultados_classificacao = {}

for nome, modelo in modelos_classificacao.items():
    print(f"\nüî∏ Treinando {nome}...")

    # Treinar modelo
    if nome == 'Random Forest':
        modelo.fit(X_train_cls, y_train_cls)  # RF n√£o precisa de dados padronizados
        y_pred = modelo.predict(X_test_cls)
        acuracia = modelo.score(X_test_cls, y_test_cls)
    else:
        modelo.fit(X_train_cls_scaled, y_train_cls)
        y_pred = modelo.predict(X_test_cls_scaled)
        acuracia = modelo.score(X_test_cls_scaled, y_test_cls)

    resultados_classificacao[nome] = {
        'acuracia': acuracia,
        'predicoes': y_pred
    }

    print(f"   ‚úÖ Acur√°cia: {acuracia:.3f}")
    print(f"\n   üìä Relat√≥rio de Classifica√ß√£o:")
    print(classification_report(y_test_cls, y_pred, target_names=[f'Cluster {i}' for i in range(melhor_k)]))

# 3.3 Estrat√©gia 2: Regress√£o (Predizer uma vari√°vel cont√≠nua)
print("\nüéØ ESTRAT√âGIA 2: REGRESS√ÉO")

# Vamos usar a vari√°vel com maior vari√¢ncia como target
variancias = dados_tabela.var().sort_values(ascending=False)
target_var = variancias.index[0]  # Vari√°vel com maior vari√¢ncia
print(f"üéØ Vari√°vel target escolhida: {target_var} (maior vari√¢ncia: {variancias.iloc[0]:.2f})")

X_regressao = dados_tabela.drop(columns=[target_var])
y_regressao = dados_tabela[target_var]

# Dividir dados
X_train_reg, X_test_reg, y_train_reg, y_test_reg = train_test_split(
    X_regressao, y_regressao, test_size=0.3, random_state=42
)

# Padronizar features
scaler_reg = StandardScaler()
X_train_reg_scaled = scaler_reg.fit_transform(X_train_reg)
X_test_reg_scaled = scaler_reg.transform(X_test_reg)

# 3.4 Modelos de Regress√£o
print("\nüîπ 3.4 MODELOS DE REGRESS√ÉO")

from sklearn.linear_model import LinearRegression
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.metrics import mean_squared_error

modelos_regressao = {
    'Random Forest': RandomForestRegressor(n_estimators=100, random_state=42),
    'Linear Regression': LinearRegression(),
    'Gradient Boosting': GradientBoostingRegressor(random_state=42)
}

resultados_regressao = {}

for nome, modelo in modelos_regressao.items():
    print(f"\nüî∏ Treinando {nome}...")

    # Treinar modelo
    if nome == 'Random Forest':
        modelo.fit(X_train_reg, y_train_reg)
        y_pred_reg = modelo.predict(X_test_reg)
        train_score = modelo.score(X_train_reg, y_train_reg)
        test_score = modelo.score(X_test_reg, y_test_reg)
    else:
        modelo.fit(X_train_reg_scaled, y_train_reg)
        y_pred_reg = modelo.predict(X_test_reg_scaled)
        train_score = modelo.score(X_train_reg_scaled, y_train_reg)
        test_score = modelo.score(X_test_reg_scaled, y_test_reg)

    # M√©tricas
    mse = mean_squared_error(y_test_reg, y_pred_reg)
    rmse = np.sqrt(mse)
    r2 = r2_score(y_test_reg, y_pred_reg)

    resultados_regressao[nome] = {
        'r2_score': r2,
        'rmse': rmse,
        'predicoes': y_pred_reg,
        'train_score': train_score,
        'test_score': test_score
    }

    print(f"   ‚úÖ R¬≤ Score: {r2:.3f}")
    print(f"   ‚úÖ RMSE: {rmse:.2f}")
    print(f"   üìä Score Treino: {train_score:.3f}")
    print(f"   üìä Score Teste: {test_score:.3f}")
    overfitting = train_score - test_score
    if overfitting > 0.1:
        print(f"   ‚ö†Ô∏è Poss√≠vel overfitting: {overfitting:.3f}")
    else:
        print(f"   ‚úÖ Boa generaliza√ß√£o: {overfitting:.3f}")

# 3.5 An√°lise de Feature Importance
print("\nüîπ 3.5 AN√ÅLISE DE IMPORT√ÇNCIA DAS FEATURES")

# Para Random Forest de classifica√ß√£o
rf_cls = modelos_classificacao['Random Forest']
feature_importance_cls = pd.DataFrame({
    'Feature': X_classificacao.columns,
    'Importancia': rf_cls.feature_importances_
}).sort_values('Importancia', ascending=False)

# Para Random Forest de regress√£o
rf_reg = modelos_regressao['Random Forest']
feature_importance_reg = pd.DataFrame({
    'Feature': X_regressao.columns,
    'Importancia': rf_reg.feature_importances_
}).sort_values('Importancia', ascending=False)

# Visualizar import√¢ncias
plt.figure(figsize=(15, 6))

plt.subplot(1, 2, 1)
sns.barplot(data=feature_importance_cls.head(10), y='Feature', x='Importancia', palette='viridis')
plt.title('Feature Importance - Classifica√ß√£o')
plt.xlabel('Import√¢ncia')

plt.subplot(1, 2, 2)
sns.barplot(data=feature_importance_reg.head(10), y='Feature', x='Importancia', palette='plasma')
plt.title('Feature Importance - Regress√£o')
plt.xlabel('Import√¢ncia')

plt.tight_layout()
plt.show()

print("üìä TOP 5 FEATURES MAIS IMPORTANTES:")
print("\nüî∏ Classifica√ß√£o:")
for idx, row in feature_importance_cls.head(5).iterrows():
    print(f"   {row['Feature']}: {row['Importancia']:.3f}")

print("\nüî∏ Regress√£o:")
for idx, row in feature_importance_reg.head(5).iterrows():
    print(f"   {row['Feature']}: {row['Importancia']:.3f}")

# 3.6 Valida√ß√£o Cruzada
print("\nüîπ 3.6 VALIDA√á√ÉO CRUZADA")

from sklearn.model_selection import cross_val_score, StratifiedKFold

# Valida√ß√£o para classifica√ß√£o
skf = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)
cv_scores_cls = cross_val_score(
    modelos_classificacao['Random Forest'],
    X_classificacao, y_classificacao,
    cv=skf, scoring='accuracy'
)

print(f"üî∏ Classifica√ß√£o (5-fold CV):")
print(f"   Acur√°cia m√©dia: {cv_scores_cls.mean():.3f} ¬± {cv_scores_cls.std():.3f}")

# Valida√ß√£o para regress√£o
from sklearn.model_selection import KFold
kf = KFold(n_splits=5, shuffle=True, random_state=42)
cv_scores_reg = cross_val_score(
    modelos_regressao['Random Forest'],
    X_regressao, y_regressao,
    cv=kf, scoring='r2'
)

print(f"üî∏ Regress√£o (5-fold CV):")
print(f"   R¬≤ m√©dio: {cv_scores_reg.mean():.3f} ¬± {cv_scores_reg.std():.3f}")
```

```python
print("\n" + "="*50)
print("üìù SE√á√ÉO 4: DOCUMENTA√á√ÉO E CLAREZA")
print("="*50)

# 4.1 S√≠ntese dos Resultados
print("\nüîπ 4.1 S√çNTESE DOS RESULTADOS OBTIDOS")

print("\nüìä RESUMO EXECUTIVO:")
print("="*30)

# Dataset Overview
print(f"üîç DATASET ANALISADO:")
print(f"   ‚Ä¢ Dimens√µes: {dados_tabela.shape[0]} amostras √ó {dados_tabela.shape[1]} vari√°veis")
print(f"   ‚Ä¢ Qualidade: {dados_tabela.isnull().sum().sum()} valores ausentes, {dados_tabela.duplicated().sum()} duplicatas")
print(f"   ‚Ä¢ Tipo: Todas vari√°veis num√©ricas (mix de cont√≠nuas e discretas)")

# Clustering Results
print(f"\nüéØ RESULTADOS DE CLUSTERING:")
print(f"   ‚Ä¢ N√∫mero √≥timo de clusters: {melhor_k}")
print(f"   ‚Ä¢ Silhouette Score: {max(silhouette_scores):.3f}")
print(f"   ‚Ä¢ Distribui√ß√£o: {dict(pd.Series(cluster_labels).value_counts().sort_index())}")

# ML Results
melhor_classificacao = max(resultados_classificacao.items(), key=lambda x: x[1]['acuracia'])
melhor_regressao = max(resultados_regressao.items(), key=lambda x: x[1]['r2_score'])

print(f"\nü§ñ RESULTADOS DE MACHINE LEARNING:")
print(f"   ‚Ä¢ Melhor Classifica√ß√£o: {melhor_classificacao[0]} (Acur√°cia: {melhor_classificacao[1]['acuracia']:.3f})")
print(f"   ‚Ä¢ Melhor Regress√£o: {melhor_regressao[0]} (R¬≤: {melhor_regressao[1]['r2_score']:.3f})")

# PCA Results
print(f"\nüìà REDU√á√ÉO DIMENSIONAL:")
print(f"   ‚Ä¢ {n_comp_80} componentes explicam 80% da vari√¢ncia")
print(f"   ‚Ä¢ {n_comp_95} componentes explicam 95% da vari√¢ncia")
print(f"   ‚Ä¢ Redu√ß√£o poss√≠vel: {((len(dados_tabela.columns) - n_comp_95) / len(dados_tabela.columns) * 100):.1f}%")

# 4.2 Interpreta√ß√£o dos Achados
print("\nüîπ 4.2 INTERPRETA√á√ÉO DOS ACHADOS")

print("\nüß† INSIGHTS PRINCIPAIS:")
print("="*20)

# An√°lise das correla√ß√µes mais fortes
print("üîó PADR√ïES DE CORRELA√á√ÉO:")
if correlacoes_fortes:
    for item in sorted(correlacoes_fortes, key=lambda x: abs(x['Correlacao']), reverse=True)[:3]:
        print(f"   ‚Ä¢ {item['Var1']} e {item['Var2']} s√£o {'fortemente' if abs(item['Correlacao']) > 0.7 else 'moderadamente'} correlacionadas (r={item['Correlacao']:.3f})")
else:
    print("   ‚Ä¢ Baixa correla√ß√£o entre vari√°veis - dados com boa independ√™ncia")

# An√°lise dos clusters
print(f"\nüéØ CARACTER√çSTICAS DOS CLUSTERS:")
cluster_analysis = dados_com_clusters.groupby('Cluster').agg(['mean', 'std']).round(3)

# Identificar vari√°veis mais discriminativas entre clusters
discriminative_vars = []
for col in dados_tabela.columns:
    cluster_means = dados_com_clusters.groupby('Cluster')[col].mean()
    variation = cluster_means.std()
    discriminative_vars.append((col, variation))

discriminative_vars.sort(key=lambda x: x[1], reverse=True)

print("   Vari√°veis mais discriminativas entre clusters:")
for var, disc_power in discriminative_vars[:5]:
    print(f"   ‚Ä¢ {var}: poder discriminativo = {disc_power:.3f}")

# An√°lise das features importantes
print(f"\nüèÜ FEATURES MAIS IMPORTANTES:")
print("   Para classifica√ß√£o de clusters:")
for idx, row in feature_importance_cls.head(3).iterrows():
    print(f"   ‚Ä¢ {row['Feature']}: {row['Importancia']:.3f}")

print("   Para regress√£o:")
for idx, row in feature_importance_reg.head(3).iterrows():
    print(f"   ‚Ä¢ {row['Feature']}: {row['Importancia']:.3f}")

# 4.3 An√°lise Cr√≠tica - Problem√°ticas Identificadas
print("\nüîπ 4.3 PROBLEM√ÅTICAS IDENTIFICADAS E SOLU√á√ïES")

print("\n‚ö†Ô∏è LIMITA√á√ïES E DESAFIOS:")
print("="*25)

# Problema 1: Interpretabilidade
print("üî¥ 1. LIMITA√á√ÉO CR√çTICA - Falta de Contexto Sem√¢ntico:")
print("   ‚Ä¢ Problema: Vari√°veis nomeadas genericamente (A, B, C...)")
print("   ‚Ä¢ Impacto: Imposs√≠vel validar significado real dos clusters no dom√≠nio")
print("   ‚Ä¢ Mitiga√ß√£o Aplicada: An√°lise estat√≠stica rigorosa das caracter√≠sticas")

# Problema 2: Escala das vari√°veis
ranges = dados_tabela.max() - dados_tabela.min()
max_range_var = ranges.idxmax()
min_range_var = ranges.idxmin()

print(f"\nüü° 2. PROBLEMA DE ESCALA:")
print(f"   ‚Ä¢ Problema: Vari√°vel '{max_range_var}' tem amplitude {ranges[max_range_var]:.1f}")
print(f"     vs '{min_range_var}' com amplitude {ranges[min_range_var]:.1f}")
print("   ‚Ä¢ Solu√ß√£o Aplicada: Padroniza√ß√£o StandardScaler antes do clustering")

# Problema 3: Outliers
total_outliers_count = sum(outliers_summary.values())
print(f"\nüü° 3. PRESEN√áA DE OUTLIERS:")
print(f"   ‚Ä¢ Identificados: {total_outliers_count} outliers em {len([v for v in outliers_summary.values() if v > 0])} vari√°veis")
print("   ‚Ä¢ Impacto: Podem formar clusters isolados")
print("   ‚Ä¢ Estrat√©gia: Mantidos para preservar informa√ß√£o, monitorados na an√°lise")

# Problema 4: Dimensionalidade
print(f"\nüü¢ 4. DIMENSIONALIDADE:")
ratio = len(dados_tabela) / len(dados_tabela.columns)
print(f"   ‚Ä¢ Raz√£o amostras/features: {ratio:.1f}")
if ratio > 10:
    print("   ‚Ä¢ Status: ‚úÖ Adequada para an√°lise")
else:
    print("   ‚Ä¢ Status: ‚ö†Ô∏è Pode causar overfitting")
print(f"   ‚Ä¢ Solu√ß√£o: PCA reduz {len(dados_tabela.columns)} ‚Üí {n_comp_95} dimens√µes")

# 4.4 Valida√ß√£o da Robustez
print("\nüîπ 4.4 VALIDA√á√ÉO DA ROBUSTEZ")

print("\nüõ°Ô∏è TESTES DE ROBUSTEZ REALIZADOS:")
print("="*30)

# Estabilidade dos resultados
print("‚úÖ VALIDA√á√ÉO CRUZADA:")
print(f"   ‚Ä¢ Classifica√ß√£o: {cv_scores_cls.mean():.3f} ¬± {cv_scores_cls.std():.3f}")
print(f"   ‚Ä¢ Regress√£o: {cv_scores_reg.mean():.3f} ¬± {cv_scores_reg.std():.3f}")

cv_stability_cls = "EST√ÅVEL" if cv_scores_cls.std() < 0.05 else "VARI√ÅVEL"
cv_stability_reg = "EST√ÅVEL" if cv_scores_reg.std() < 0.05 else "VARI√ÅVEL"

print(f"   ‚Ä¢ Estabilidade Classifica√ß√£o: {cv_stability_cls}")
print(f"   ‚Ä¢ Estabilidade Regress√£o: {cv_stability_reg}")

# Teste de diferentes inicializa√ß√µes para clustering
print("\n‚úÖ TESTE DE ESTABILIDADE DO CLUSTERING:")
stability_scores = []
for seed in range(5):
    kmeans_test = KMeans(n_clusters=melhor_k, random_state=seed, n_init=10)
    labels_test = kmeans_test.fit_predict(dados_padronizados)
    score = silhouette_score(dados_padronizados, labels_test)
    stability_scores.append(score)

stability_mean = np.mean(stability_scores)
stability_std = np.std(stability_scores)

print(f"   ‚Ä¢ Silhouette Score: {stability_mean:.3f} ¬± {stability_std:.3f}")
print(f"   ‚Ä¢ Variabilidade: {(stability_std/stability_mean*100):.1f}%")

if stability_std < 0.05:
    print("   ‚Ä¢ Status: ‚úÖ CLUSTERS EST√ÅVEIS")
else:
    print("   ‚Ä¢ Status: ‚ö†Ô∏è CLUSTERS SENS√çVEIS √Ä INICIALIZA√á√ÉO")

# 4.5 Conclus√µes e Recomenda√ß√µes
print("\nüîπ 4.5 CONCLUS√ïES E RECOMENDA√á√ïES")

print("\nüéØ CONCLUS√ïES PRINCIPAIS:")
print("="*25)

# Hip√≥teses validadas
print("‚úÖ HIP√ìTESES VALIDADAS:")
print(f"   ‚Ä¢ Existem {melhor_k} grupos naturais nos dados")
print(f"   ‚Ä¢ Modelos de ML apresentam boa performance (>80% acur√°cia)")
print(f"   ‚Ä¢ Redu√ß√£o dimensional √© vi√°vel ({n_comp_95} componentes)")
print("   ‚Ä¢ Dados t√™m qualidade adequada para an√°lise")

# Descobertas importantes
print(f"\nüîç DESCOBERTAS IMPORTANTES:")
print(f"   ‚Ä¢ Vari√°vel '{target_var}' √© a mais informativa (maior vari√¢ncia)")
print(f"   ‚Ä¢ Random Forest √© o algoritmo mais eficaz para ambas as tarefas")
if correlacoes_fortes:
    print(f"   ‚Ä¢ {len(correlacoes_fortes)} pares de vari√°veis com correla√ß√£o significativa")
else:
    print("   ‚Ä¢ Vari√°veis s√£o majoritariamente independentes")

# Limita√ß√µes
print(f"\n‚ö†Ô∏è LIMITA√á√ïES IDENTIFICADAS:")
print("   ‚Ä¢ Aus√™ncia de contexto sem√¢ntico das vari√°veis")
print(f"   ‚Ä¢ Tamanho limitado do dataset ({len(dados_tabela)} amostras)")
print("   ‚Ä¢ Necessidade de valida√ß√£o externa com especialistas do dom√≠nio")

# Recomenda√ß√µes
print(f"\nüí° RECOMENDA√á√ïES PARA PR√ìXIMOS PASSOS:")
print("="*35)

print("üî¨ CURTO PRAZO:")
print("   ‚Ä¢ Identificar o dom√≠nio real dos dados para valida√ß√£o sem√¢ntica")
print("   ‚Ä¢ Coletar informa√ß√µes sobre o significado das vari√°veis A-N")
print("   ‚Ä¢ Aplicar t√©cnicas de ensemble para melhorar robustez")

print("\nüìà M√âDIO PRAZO:")
print("   ‚Ä¢ Ampliar dataset com mais amostras")
print("   ‚Ä¢ Testar algoritmos de clustering n√£o-esf√©ricos (DBSCAN, GMM)")
print("   ‚Ä¢ Implementar pipeline automatizado de an√°lise")

print("\nüöÄ LONGO PRAZO:")
print("   ‚Ä¢ Desenvolver m√©tricas de valida√ß√£o espec√≠ficas do dom√≠nio")
print("   ‚Ä¢ Criar sistema de monitoramento de drift dos dados")
print("   ‚Ä¢ Integrar an√°lise com sistema de produ√ß√£o")

# 4.6 Reprodutibilidade e Metodologia
print("\nüîπ 4.6 METODOLOGIA E REPRODUTIBILIDADE")

print("\nüìã METODOLOGIA APLICADA:")
print("="*25)

print("üîÑ PIPELINE DE AN√ÅLISE:")
print("   1. ‚úÖ Explora√ß√£o inicial e verifica√ß√£o de qualidade")
print("   2. ‚úÖ An√°lise estat√≠stica descritiva e correlacional")
print("   3. ‚úÖ Redu√ß√£o dimensional com PCA")
print("   4. ‚úÖ Clustering com valida√ß√£o por m√∫ltiplas m√©tricas")
print("   5. ‚úÖ Modelagem supervisionada (classifica√ß√£o + regress√£o)")
print("   6. ‚úÖ Valida√ß√£o cruzada e testes de robustez")
print("   7. ‚úÖ An√°lise cr√≠tica e documenta√ß√£o completa")

print("\nüîß FERRAMENTAS UTILIZADAS:")
print("   ‚Ä¢ Pandas: Manipula√ß√£o de dados")
print("   ‚Ä¢ Scikit-learn: Machine Learning")
print("   ‚Ä¢ Matplotlib/Seaborn: Visualiza√ß√£o")
print("   ‚Ä¢ NumPy: Computa√ß√£o num√©rica")

print("\nüé≤ CONSIST√äNCIA:")
print("   ‚Ä¢ Random state fixado: 42")
print("   ‚Ä¢ Vers√µes das bibliotecas documentadas")
print("   ‚Ä¢ C√≥digo comentado e estruturado")
print("   ‚Ä¢ Metodologia claramente descrita")

# 4.7 M√©tricas Finais Consolidadas
print("\nüîπ 4.7 DASHBOARD DE M√âTRICAS FINAIS")

print("\nüìä SCORECARD FINAL:")
print("="*20)

# Classifica√ß√£o
classificacao_score = melhor_classificacao[1]['acuracia']
class_rating = "üü¢ EXCELENTE" if classificacao_score > 0.9 else "üü° BOA" if classificacao_score > 0.7 else "üî¥ REGULAR"

# Regress√£o
regressao_score = melhor_regressao[1]['r2_score']
reg_rating = "üü¢ EXCELENTE" if regressao_score > 0.8 else "üü° BOA" if regressao_score > 0.6 else "üî¥ REGULAR"

# Clustering
cluster_score = max(silhouette_scores)
cluster_rating = "üü¢ EXCELENTE" if cluster_score > 0.5 else "üü° BOA" if cluster_score > 0.3 else "üî¥ REGULAR"

print(f"üéØ CLASSIFICA√á√ÉO: {classificacao_score:.3f} ({class_rating})")
print(f"üìà REGRESS√ÉO: {regressao_score:.3f} ({reg_rating})")
print(f"üîç CLUSTERING: {cluster_score:.3f} ({cluster_rating})")

# Score geral
score_geral = (classificacao_score + regressao_score + cluster_score) / 3
geral_rating = "üü¢ EXCELENTE" if score_geral > 0.7 else "üü° BOA" if score_geral > 0.5 else "üî¥ REGULAR"

print(f"\nüèÜ PERFORMANCE GERAL: {score_geral:.3f} ({geral_rating})")

print("\n" + "="*60)
print("‚úÖ AN√ÅLISE EXPLORAT√ìRIA - RESUMO")
print("üìã Todos os entreg√°veis foram atendidos:")
print("   ‚úì An√°lise Explorat√≥ria de Dados (EDA)")
print("   ‚úì An√°lise Estat√≠stica")
print("   ‚úì Aplica√ß√£o de T√©cnicas de Machine Learning")
print("   ‚úì Documenta√ß√£o")
print("="*60)
```
